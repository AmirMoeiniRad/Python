# Seaborn Linear Regression Plot

```python
import seaborn as sns

# Display plots inside notebook
%matplotlib inline

# One of built-in Seaborn datasets.
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### **seaborn.lmplot** only plots the regression model. It does not return the regression parameters.

```python
# Linear Regression
# x: the data for the x axis
# y: the data for the y axis
# aspect: the ratio of plot width to plot height
# data: the source of the data to be plotted. Here: the pandas 'tips' datagram
sns.lmplot(x='total_bill',y='tip',data=tips,aspect=2,markers='o')
```

```python
# Separate the points based on the 'sex' column and use a different color for each category (male or female): hue='sex'
# Now, a separate regression line is calculated for each of the 'male' and 'female' categories.
sns.lmplot(x='total_bill',y='tip',data=tips,aspect=2,markers='o',hue='sex',legend=True)
```
