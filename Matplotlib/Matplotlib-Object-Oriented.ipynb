# Matplotlib's Object-Oriented Method

```python
import matplotlib.pyplot as plt
import numpy as np

# sample data
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
# Create a figure (an empty canvas)
# Pay attention to 'figsize' and 'dpi' params. Figsize define the width-to-height aspect ratio.
fig = plt.figure(figsize=(12,6), dpi=200)

# Add two sets of axes to the figure
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])    # main axes: left, bottom, width, height (range 0 to 1)
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])    # inset axes

# Plot on that set of axes. Larger figure.
axes1.plot(x, y, 'b',marker='o',markerfacecolor="#FFFF00",markeredgecolor="black",markeredgewidth=2,label='$y=x^2$')    # 'b' --> blue
axes1.set_xlabel('X Values')
axes1.set_ylabel('Y Values')
axes1.set_title('Title1')
axes1.legend()
axes1.text(1, 3, r"$y=x^2$", fontsize=20, color="blue")

# Inset figure
axes2.plot(y, x, 'r--',alpha=0.5,linewidth=0.75,marker='*',markersize=5,label='$x=y^2$')    # 'r--' --> red dotted line. 'alpha' --> transparency
axes2.set_xlabel('X Values')
axes2.set_ylabel('Y Values')
axes2.set_title('Title2');
axes2.legend(loc=4)    # lower right corner
axes2.set_ylim([-1, 6])    # y axis range
axes2.set_xlim([-1, 26])   # x axis range
axes2.grid(alpha=0.5, linestyle='dashed')
```
