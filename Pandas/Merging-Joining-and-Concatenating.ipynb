# Merging, Joining, and Concatenating Pandas DataFrames

These ways of combining DataFrames together can sometine be a little bit confusing!

```python
import pandas as pd
```

## (1) Concatenation

Concatenation basically glues together DataFrames. Keep in mind that dimensions should match along the axis you are concatenating on.

### Example DataFrames

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7]) 
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
# Default along columns (axis=1). In other words, put rows beside one another.
# The number of rows equals the sum of rows of the three original dataframes: 12
pd.concat([df1,df2,df3])
```

```python
# Combine dataframes along the rows. In other words, put columns beside one another.
# The number of rows & columns equals the sum of rows & columns of the original dataframes: (12,12)
pd.concat([df1,df2,df3],axis=1)
```

## (2) Merging

The merge function depends on common column(s) in both dataframes.

### Example DataFrames

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})    
```

```python
left
```

```python
right
```

```python
# how='inner' --> The common/similar column 'key' lies before the two dataframes.
pd.merge(left,right,how='inner',on='key')
```

### A More Complicated Example

```python
L = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
R = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
L
```

```python
R
```

```python
# 'inner' only uses similar 'key1' & 'key2' elements between the two dataframes to merge them.
pd.merge(L, R, how='inner', on=['key1', 'key2'])
```

```python
# 'outer' uses both similar and different 'key1' & 'key2' elements between the two dataframes to merge them.
pd.merge(L, R, how='outer', on=['key1', 'key2'])
```

## (3) Joining
Joining combines the columns of two differently-indexed DataFrames into a single DataFrame.

```python
LL = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

RR = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
LL
```

```python
RR
```

```python
# Here, unlike the 'merging' example, there are no similar columns in the two dataframes. 'Joining' acts based on similar indices with how='inner'.
LL.join(RR)
```

```python
# Here, 'Joining' acts based on both similar and different indices with how='outer'.
LL.join(RR, how='outer')
```
